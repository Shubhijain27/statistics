1. What is hypothesis testing in statistics?


---


Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It involves formulating two competing hypotheses—the null and alternative hypotheses—and using sample data to determine which hypothesis is more likely to be true.


---



2. What is the null hypothesis, and how does it differ from the alternative hypothesis?


---


The null hypothesis (H₀) posits that there is no effect or difference in the population, serving as the default assumption. In contrast, the alternative hypothesis (H₁ or Ha) suggests that there is an effect or difference. Hypothesis testing evaluates whether the sample data provide sufficient evidence to reject H₀ in favor of H₁.


---



3. What is the significance level in hypothesis testing, and why is it important?


---


The significance level (α) is the threshold for determining whether a result is statistically significant. Commonly set at 0.05, it represents a 5% risk of rejecting the null hypothesis when it is actually true (Type I error). Choosing an appropriate α helps balance the risks of Type I and Type II errors.

4. What does a P-value represent in hypothesis testing?


---


A P-value quantifies the probability of obtaining results at least as extreme as those observed, assuming the null hypothesis is true. A smaller P-value indicates stronger evidence against H₀, suggesting that the observed data are unlikely under the null hypothesis.


---



5. How do you interpret the P-value in hypothesis testing?


---


If P-value ≤ α: Reject the null hypothesis; the result is statistically significant.

If P-value > α: Fail to reject the null hypothesis; the result is not statistically significant.

This interpretation helps determine whether the observed data provide sufficient evidence to support the alternative hypothesis.

6. What are Type I and Type II errors in hypothesis testing?


---



Type I Error (False Positive): Incorrectly rejecting a true null hypothesis.


Type II Error (False Negative): Failing to reject a false null hypothesis.

Balancing these errors is crucial in hypothesis testing, often by selecting an appropriate significance level and ensuring adequate sample size.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
A Z-test is used when the population standard deviation is known, and the sample size is large (typically n ≥ 30).

```python
import numpy as np
from scipy import stats

# Sample data
sample = [105, 110, 98, 100, 102, 108, 95, 99, 101, 104]
sample_mean = np.mean(sample)
population_mean = 100
population_std = 15
n = len(sample)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot
Visualizing the decision regions helps in understanding the critical areas where we would reject the null hypothesis.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters
population_mean = 100
population_std = 15
n = 30
alpha = 0.05

# Sample data
sample = np.random.normal(loc=105, scale=15, size=n)
sample_mean = np.mean(sample)

# Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Critical Z-values
z_critical = stats.norm.ppf(1 - alpha/2)

# Plot
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y)
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5)
plt.axvline(z_score, color='black', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.title('Two-tailed Z-test Decision Regions')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

5. Create a Python function that calculates and visualizes Type I and Type II errors during hypothesis testing
Type I error occurs when we reject a true null hypothesis, while Type II error occurs when we fail to reject a false null hypothesis

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(mu0, mu1, sigma, n, alpha=0.05):
    # Calculate standard error
    se = sigma / np.sqrt(n)

    # Critical value for Type I error
    z_alpha = norm.ppf(1 - alpha)
    x_critical = mu0 + z_alpha * se

    # Type II error: beta
    beta = norm.cdf((x_critical - mu1) / se)

    # Plot distributions
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.plot(x, y0, label='H0: μ = μ0')
    plt.plot(x, y1, label='H1: μ = μ1')
    plt.axvline(x_critical, color='red', linestyle='--', label='Critical Value')
    plt.fill_between(x, 0, y0, where=(x >= x_critical), color='red', alpha=0.3, label='Type I Error')
    plt.fill_between(x, 0, y1, where=(x < x_critical), color='blue', alpha=0.3, label='Type II Error')
    plt.legend()
    plt.title('Type I and Type II Errors')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.show()

    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage
visualize_errors(mu0=100, mu1=105, sigma=15, n=30, alpha=0.05)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
A confidence interval provides a range of values within which we can be confident that the population parameter lies.



```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    sem = stats.sem(data)
    margin = sem * stats.t.ppf((1 + confidence) / 2., n-1)
    return mean - margin, mean + margin

# Example usage
sample = [100, 102, 98, 101, 99]
ci_lower, ci_upper = confidence_interval(sample)
print(f"95% confidence interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

Q14: How to perform a goodness-of-fit test in Python to compare observed data to an expected distribution?
A5:

The Chi-square goodness-of-fit test determines whether a sample matches an expected distribution. It's used to test if the observed frequencies differ from the expected frequencies.

```python
import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Step 1: Observed frequencies (e.g., categories A–E)
observed = np.array([25, 30, 20, 15, 10])

# Step 2: Expected frequencies (uniform distribution in this case)
total = observed.sum()
expected = np.full_like(observed, fill_value=total / len(observed))

# Step 3: Chi-square test
chi_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Display results
print("Chi-Square Goodness-of-Fit Test")
print("--------------------------------")
print(f"Observed: {observed}")
print(f"Expected: {expected}")
print(f"Chi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Conclusion: Significant difference from expected distribution.")
else:
    print("Conclusion: No significant difference from expected distribution.")

# Step 5: Visualize observed vs expected
labels = [f'Cat {i+1}' for i in range(len(observed))]
x = np.arange(len(observed))
width = 0.35

plt.bar(x - width/2, observed, width=width, label='Observed')
plt.bar(x + width/2, expected, width=width, label='Expected', alpha=0.7)
plt.xticks(x, labels)
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies')
plt.legend()
plt.grid(axis='y')
plt.show()
```

Q16: How to implement an F-test in Python to compare the variances of two random samples?
A7:

An F-test compares the variances of two populations to assess if they are significantly different. The test statistic is the ratio of the two sample variances.

```python
import numpy as np
from scipy.stats import f

# Sample data
sample1 = [10, 12, 14, 16, 18]
sample2 = [20, 22, 24, 26, 28]

# Calculate variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic
F = var1 / var2

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Compute p-value
p_value = 1 - f.cdf(F, df1, df2)

print(f"F-statistic: {F:.2f}")
print(f"p-value: {p_value:.4f}")
```

18 Perform a one-way ANOVA test using Python to compare the means of different groups and plot the resultsD

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for 3 groups
data = {
    'Score': [88, 92, 85, 90, 87, 78, 75, 80, 82, 77, 92, 95, 91, 93, 94],
    'Group': ['A']*5 + ['B']*5 + ['C']*5
}
df = pd.DataFrame(data)

# One-way ANOVA
grouped = df.groupby('Group')['Score'].apply(list)
f_stat, p_val = stats.f_oneway(*grouped)

print("One-Way ANOVA Results")
print("----------------------")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Conclusion: Significant differences between group means.")
else:
    print("Conclusion: No significant difference between group means.")

# Plotting boxplot
sns.boxplot(x='Group', y='Score', data=df)
plt.title("Group Score Comparison")
plt.show()
```

19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
from scipy.stats import shapiro, levene
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan

def check_anova_assumptions(df, group_col='Group', value_col='Score'):
    print("\nANOVA Assumptions Check")
    print("-----------------------")

    # 1. Normality - Shapiro-Wilk for each group
    for grp, values in df.groupby(group_col)[value_col]:
        stat, p = shapiro(values)
        print(f"Shapiro-Wilk ({grp}): p = {p:.4f}")

    # 2. Homogeneity of variances - Levene’s Test
    grouped = df.groupby(group_col)[value_col].apply(list)
    lev_stat, lev_p = levene(*grouped)
    print(f"Levene's Test for equal variances: p = {lev_p:.4f}")

    # 3. Independence - Residuals via Breusch-Pagan test
    model = ols(f"{value_col} ~ C({group_col})", data=df).fit()
    _, pval, _, _ = het_breuschpagan(model.resid, model.model.exog)
    print(f"Breusch-Pagan test (residual independence): p = {pval:.4f}")

# Example usage
check_anova_assumptions(df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import statsmodels.api as sm

# Sample dataset
df2 = pd.DataFrame({
    'Score': [23, 25, 24, 26, 22, 31, 29, 30, 32, 33,
              35, 34, 36, 38, 37, 41, 42, 40, 43, 44],
    'FactorA': ['Low']*10 + ['High']*10,
    'FactorB': ['X']*5 + ['Y']*5 + ['X']*5 + ['Y']*5
})

# Two-way ANOVA
model = ols('Score ~ C(FactorA) + C(FactorB) + C(FactorA):C(FactorB)', data=df2).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nTwo-Way ANOVA Results")
print(anova_table)

# Interaction Plot
sns.pointplot(x='FactorB', y='Score', hue='FactorA', data=df2, capsize=0.1)
plt.title("Interaction Plot: FactorA vs FactorB")
plt.show()
```

21 - Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Parameters for F-distribution
dfn = 2  # degrees of freedom numerator
dfd = 10  # degrees of freedom denominator
x = np.linspace(0, 5, 500)
y = f.pdf(x, dfn, dfd)

# Plotting
plt.plot(x, y, 'b-', label=f'F-distribution (dfn={dfn}, dfd={dfd})')
plt.fill_between(x, y, where=(x > f.ppf(0.95, dfn, dfd)), color='red', alpha=0.3, label='Critical Region (α=0.05)')
plt.title("F-distribution with α = 0.05")
plt.xlabel('F-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Explanation:
print("""
The F-distribution is used in ANOVA to compare group variances by analyzing the ratio of between-group
variance to within-group variance. If the F-statistic falls in the critical region, we reject the null hypothesis
that all group means are equal.
""")
```

22 -  Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group meansD

```python
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'Score': [85, 87, 90, 88, 86,   # Group A
              78, 76, 80, 79, 77,   # Group B
              92, 95, 94, 93, 96],  # Group C
    'Group': ['A']*5 + ['B']*5 + ['C']*5
}
df = pd.DataFrame(data)

# Perform one-way ANOVA
grouped_data = df.groupby('Group')['Score'].apply(list)
f_stat, p_value = stats.f_oneway(*grouped_data)

# Print ANOVA result
print("One-Way ANOVA Results")
print("----------------------")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusion: Significant differences exist between at least two group means.")
else:
    print("Conclusion: No significant difference between group means.")

# Visualization with boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Group', y='Score', data=df)
plt.title('Boxplot of Scores by Group')
plt.xlabel('Group')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

23 - Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Simulate data from a normal distribution
sample1 = np.random.normal(loc=50, scale=5, size=100)  # mean=50, std=5
sample2 = np.random.normal(loc=52, scale=5, size=100)  # slightly different mean

# Histogram for visualization
plt.hist(sample1, bins=15, alpha=0.6, label='Sample 1 (mean ~ 50)')
plt.hist(sample2, bins=15, alpha=0.6, label='Sample 2 (mean ~ 52)')
plt.title("Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

# 1. One-sample t-test: is sample1's mean significantly different from 50?
t_stat1, p_val1 = stats.ttest_1samp(sample1, popmean=50)
print("One-Sample t-test (μ = 50)")
print(f"T-statistic: {t_stat1:.4f}, P-value: {p_val1:.4f}")
if p_val1 < 0.05:
    print("Conclusion: The sample mean is significantly different from 50.\n")
else:
    print("Conclusion: No significant difference from mean = 50.\n")

# 2. Two-sample t-test: are sample1 and sample2 significantly different?
t_stat2, p_val2 = stats.ttest_ind(sample1, sample2, equal_var=True)
print("Two-Sample t-test")
print(f"T-statistic: {t_stat2:.4f}, P-value: {p_val2:.4f}")
if p_val2 < 0.05:
    print("Conclusion: The two sample means are significantly different.")
else:
    print("Conclusion: No significant difference between the two sample means.")
```

24 - Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Step 1: Simulate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50

# Step 2: Visualize the data
plt.hist(data, bins=10, edgecolor='black', alpha=0.7)
plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label=f"Mean = {np.mean(data):.2f}")
plt.title("Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.show()

# Step 3: Perform one-sample t-test
# H0: mean = 100
# H1: mean ≠ 100
t_stat, p_val = stats.ttest_1samp(data, popmean=100)

# Step 4: Report results
print("One-Sample t-Test Results")
print("--------------------------")
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Conclusion: Reject the null hypothesis. The mean is significantly different from 100.")
else:
    print("Conclusion: Fail to reject the null hypothesis. No significant difference from 100.")
```

25 - Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
from statsmodels.stats.proportion import proportions_ztest

# Sample data
successes = [45, 30]   # number of successes in each group
nobs = [100, 100]      # total observations in each group

# Perform Z-test
z_stat, p_val = proportions_ztest(successes, nobs)

# Results
print("Z-Test for Comparing Proportions")
print("---------------------------------")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Conclusion: Significant difference between proportions.")
else:
    print("Conclusion: No significant difference between proportions.")
```

26 - Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results




```python
import numpy as np
import matplotlib.pyplot as plt

# Generate two normal samples
np.random.seed(1)
data1 = np.random.normal(0, 10, 100)
data2 = np.random.normal(0, 15, 100)

# Compute variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Perform F-test (larger variance in numerator)
f_stat = var1 / var2
df1 = len(data1) - 1
df2 = len(data2) - 1

from scipy.stats import f
p_val = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))

print("\nF-Test for Variance Comparison")
print("-------------------------------")
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Conclusion: Variances are significantly different.")
else:
    print("Conclusion: No significant difference in variances.")

# Visualize the distributions
plt.boxplot([data1, data2], labels=['Data1', 'Data2'])
plt.title('Boxplot Comparison of Two Datasets')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

27 -  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
from scipy.stats import chisquare

# Simulated observed data
observed = [18, 22, 20, 25, 15]

# Expected frequencies (uniform distribution)
expected = [20, 20, 20, 20, 20]

# Perform chi-square test
chi_stat, p_val = chisquare(f_obs=observed, f_exp=expected)

print("\nChi-Square Goodness of Fit Test")
print("--------------------------------")
print(f"Chi-square statistic: {chi_stat:.4f}")
print(f"P-value: {p_val:.4f}")
if p_val < 0.05:
    print("Conclusion: Observed frequencies differ significantly from expected.")
else:
    print("Conclusion: No significant difference from expected frequencies.")
```
